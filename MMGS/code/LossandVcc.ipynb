```python
import re
import matplotlib.pyplot as plt

# ========= 使用者可調參數 =========
log_path = r"C:\Users\User\Desktop\mmgs_method1_trainlog.txt"  # log 檔路徑
out_loss = "method1_train_loss.png"
out_acc  = "method1_train_acc.png"

PRINT_EVERY = 5        # 🔧 每幾回合有印 log（例如 5、10、20）
TOTAL_EPOCHS = 1200    # 總訓練回合（僅作說明，不影響繪圖）
# =================================

epochs, losses, accs = [], [], []

pattern = re.compile(
    r"\[Epoch\s+(\d+)\]\s+🔧\s+LR:\s+([0-9.]+)\s+\|\s+AvgLoss:\s+([0-9.]+)\s+\|\s+Train Acc:\s+([0-9.]+)%"
)

# === 解析 log ===
with open(log_path, "r", encoding="utf-8") as f:
    for line in f:
        m = pattern.search(line)
        if m:
            ep = int(m.group(1))
            if ep % PRINT_EVERY != 0:
                continue
            loss = float(m.group(3))
            acc = float(m.group(4))
            epochs.append(ep)
            losses.append(loss)
            accs.append(acc)

if not epochs:
    raise RuntimeError("❌ 沒有解析到資料，請確認 log 檔與 PRINT_EVERY 設定")

# =======================
# 📉 Training Loss Figure
# =======================
plt.figure(figsize=(7, 4))
plt.plot(epochs, losses, linewidth=2)
plt.xlabel("Epoch")
plt.ylabel("Loss Value")
plt.title("MT-GSGCN Loss")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(out_loss, dpi=300)
plt.show()
plt.close()

# ==========================
# 📈 Training Accuracy Figure
# ==========================
plt.figure(figsize=(7, 4))
plt.plot(epochs, accs, linewidth=2)
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("MT-GSGCN Accuracy")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(out_acc, dpi=300)
plt.show()
plt.close()

print("✅ 圖片已輸出完成：")
print(" -", out_loss)
print(" -", out_acc)
```
