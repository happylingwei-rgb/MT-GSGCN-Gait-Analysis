```python
# -*- coding: utf-8 -*-
# ✅ MMGS 固定抽樣版（整個訓練開始時抽一次 → 每個 epoch 用相同片段）
#    - 保留預處理開關（平均中心化、關節標準化、時間遮罩）
#    - 1200 epochs 訓練，StepLR 每 400 epoch 衰減
#    - 測試 90 次抽樣，回報 best/avg/F1 與混淆矩陣
#    - 額外列出「最佳那次」的誤判清單（person_id / 檔名 / 真實步態 → 預測步態）
#    - 🔁 第 2~50 次：接續「前一輪」最佳權重 last_run_best.pt 繼續訓練
#    - 仍會追蹤全局最佳（Across 50 repeats）並存為 best_model_fixed.pt
#    - 🔥 新增：50 次「平均/最大/最小/變異數」統計（avg_acc_90、best_run_acc、best_macro_f1），寫入 Excel 的 summary 分頁

import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from torch.utils.data import DataLoader, TensorDataset
from collections import Counter
from sklearn.metrics import confusion_matrix, f1_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from datetime import datetime
from copy import deepcopy  # 用於保存每輪最佳模型權重

from net.mt_gsgcn_model import Model  # 你的 AGS-GCN 模型（預期輸入 [N, C, V, T, M]）
#from net.st_gcn import Model

# === 基本設定 ===
data_dir = '../datasets/database_joints'
selected_joints = [18, 17, 16, 0, 12, 13, 14]   # 7 個下肢關節 + SpineBase
time_steps = 48
num_channels = 3
test_ids = [2, 20, 6, 11, 12, 13, 18, 26]
label_mapping = {'knee': 0, 'normal': 1, 'padding': 2}
num_classes = 3

# === 預處理開關 ===
USE_MEAN_CENTERING = False
USE_JOINT_STANDARDIZATION = False
USE_TEMPORAL_MASKING = False  # 若要完全固定資料，建議關閉

# === 隨機種子（可重現） ===
def set_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# === 收集路徑 ===
def collect_paths():
    train_paths, train_labels, test_paths, test_labels = [], [], [], []
    for class_name, label in label_mapping.items():
        class_dir = os.path.join(data_dir, class_name)
        if not os.path.isdir(class_dir):
            continue
        for person_id in os.listdir(class_dir):
            person_path = os.path.join(class_dir, person_id)
            if not os.path.isdir(person_path):
                continue
            try:
                pid = int(person_id)
            except:
                continue
            for file in os.listdir(person_path):
                if not file.endswith('.txt'):
                    continue
                fpath = os.path.join(person_path, file)
                if pid in test_ids:
                    test_paths.append(fpath)
                    test_labels.append(label)
                else:
                    train_paths.append(fpath)
                    train_labels.append(label)
    return train_paths, train_labels, test_paths, test_labels

# === 抽取有效片段（預先掃描各檔可用的連續 48 幀起點） ===
def preparse_valid_segments(paths):
    cache = {}
    for path in tqdm(paths, desc="🔍 預處理 valid 片段"):
        try:
            raw_lines = []
            with open(path, 'r') as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) == 202:
                        raw_lines.append(parts)
            if len(raw_lines) < time_steps:
                continue
            data = pd.DataFrame(raw_lines).astype(float)
            # joint state 在每關節 8 欄的第 3 欄（index 2），總共 25 關節 → 2::8
            joint_states = data.iloc[:, 2::8].values  # [frames, 25]
            valid_mask = np.any((joint_states == 1) | (joint_states == 2), axis=1)

            valid_starts = []
            count = 0
            for i in range(len(valid_mask)):
                if valid_mask[i]:
                    count += 1
                    if count >= time_steps:
                        valid_starts.append(i - (time_steps - 1))
                        count -= 1  # 允許滑動一格以取得多段
                else:
                    count = 0

            if len(valid_starts) > 0:
                cache[path] = (data, valid_starts)
        except Exception as e:
            print(f"❌ 錯誤讀取 {path}：{e}")
    return cache

# === 小工具：由路徑解析 person_id（路徑格式：.../<class>/<person_id>/<file>.txt） ===
def parse_person_id(path: str):
    parts = path.replace('\\', '/').split('/')
    try:
        return int(parts[-2])
    except:
        return parts[-2]  # 若非純數字則原樣回傳

# 反向 label 對照（0→'genou', ...）
inv_label_mapping = {v: k for k, v in label_mapping.items()}

# === 訓練資料：一次性固定抽樣，整個訓練過程都用同一批 ===
def make_fixed_train_loader(paths, labels, cache, clips_per_file=1, batch_size=32, seed=None):
    if seed is not None:
        rng = np.random.default_rng(seed)
        np.random.seed(seed)
    else:
        rng = np.random.default_rng()

    X, y = [], []
    for path, label in zip(paths, labels):
        if path not in cache:
            continue
        data, valid_starts = cache[path]
        if len(valid_starts) == 0:
            continue

        replace_flag = len(valid_starts) < clips_per_file
        chosen_starts = rng.choice(valid_starts, size=clips_per_file, replace=replace_flag)

        # 取 XYZ 座標
        joints_xyz = np.zeros((data.shape[0], len(selected_joints), num_channels), dtype=np.float32)
        for i, j in enumerate(selected_joints):
            joints_xyz[:, i, 0] = data.iloc[:, 3 + j * 8]  # X
            joints_xyz[:, i, 1] = data.iloc[:, 4 + j * 8]  # Y
            joints_xyz[:, i, 2] = data.iloc[:, 5 + j * 8]  # Z

        for s in chosen_starts:
            clip = joints_xyz[s:s + time_steps]  # [T, V, C]
            if clip.shape[0] < time_steps:
                continue

            # === 預處理 ===
            if USE_MEAN_CENTERING:
                clip = clip - clip.mean(axis=1, keepdims=True)  # 每幀以關節平均為中心
            if USE_JOINT_STANDARDIZATION:
                mean = clip.mean(axis=(0, 2), keepdims=True)
                std = clip.std(axis=(0, 2), keepdims=True) + 1e-5
                clip = (clip - mean) / std
            if USE_TEMPORAL_MASKING:
                if np.random.rand() < 0.5:
                    mask_joint = np.random.choice(clip.shape[1], size=2, replace=False)
                    clip[:, mask_joint, :] = 0
                if np.random.rand() < 0.5:
                    mask_frame = np.random.choice(clip.shape[0], size=4, replace=False)
                    clip[mask_frame, :, :] = 0

            X.append(clip)
            y.append(label)

    if not X:
        return None

    # 轉為 [N, C, V, T]
    X = np.array(X, dtype=np.float32).transpose(0, 3, 1, 2)
    y = np.array(y, dtype=np.int64)
    ds = TensorDataset(torch.tensor(X, dtype=torch.float32),
                       torch.tensor(y, dtype=torch.long))
    loader = DataLoader(ds, batch_size=batch_size, shuffle=True)
    return loader

# === 測試資料：每次 evaluate 抽一次（回傳 DataLoader + metas）===
def sample_test_once(paths, labels, cache):
    X, y = [], []
    metas = []  # 每個 sample 的資訊（path, person_id, true_label_name）

    for path, label in zip(paths, labels):
        if path not in cache:
            continue
        data, valid_starts = cache[path]
        if not valid_starts:
            continue

        start = np.random.choice(valid_starts)
        joints_xyz = np.zeros((data.shape[0], len(selected_joints), num_channels), dtype=np.float32)
        for i, j in enumerate(selected_joints):
            joints_xyz[:, i, 0] = data.iloc[:, 3 + j * 8]
            joints_xyz[:, i, 1] = data.iloc[:, 4 + j * 8]
            joints_xyz[:, i, 2] = data.iloc[:, 5 + j * 8]

        clip = joints_xyz[start:start + time_steps]  # [T, V, C]
        if clip.shape[0] < time_steps:
            continue

        if USE_MEAN_CENTERING:
            clip = clip - clip.mean(axis=1, keepdims=True)
        if USE_JOINT_STANDARDIZATION:
            mean = clip.mean(axis=(0, 2), keepdims=True)
            std = clip.std(axis=(0, 2), keepdims=True) + 1e-5
            clip = (clip - mean) / std

        X.append(clip)
        y.append(label)
        metas.append({
            "path": path,
            "person_id": parse_person_id(path),
            "true_label": label,
            "true_name": inv_label_mapping[label]
        })

    if not X:
        return None, None

    X = np.array(X, dtype=np.float32).transpose(0, 3, 1, 2)  # [N, C, V, T]
    y = np.array(y, dtype=np.int64)
    loader = DataLoader(TensorDataset(torch.tensor(X, dtype=torch.float32),
                                      torch.tensor(y, dtype=torch.long)),
                        batch_size=32, shuffle=False)
    return loader, metas

# === 建立模型 ===
def build_ags_gcn():
    # 模型期望輸入 [N, C, V, T, M]；這裡在 forward 前加一個 M 維（M=1）
    return Model(
        in_channels=3,
        num_class=num_classes,
        graph_args={'layout': 'custom_7joints_spine', 'strategy': 'spatial'},
        edge_importance_weighting=True
    )

# === 測試 90 次抽樣（額外輸出最佳那次的誤判名單）===
def evaluate_model_90_times(model, test_paths, test_labels, test_cache, device):
    model.eval()
    all_accs, all_preds_list, all_trues_list = [], [], []
    all_metas_list = []

    for i in range(90):
        preds_this_run, trues_this_run = [], []
        test_loader, metas = sample_test_once(test_paths, test_labels, test_cache)
        if test_loader is None:
            continue
        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_in = X_batch.to(device).unsqueeze(-1)  # [N, C, V, T] → [N, C, V, T, 1]
                logits = model(X_in)
                preds = logits.argmax(dim=1)
                preds_np = preds.detach().cpu().numpy()
                trues_np = y_batch.numpy()
                preds_this_run.extend(preds_np.tolist())
                trues_this_run.extend(trues_np.tolist())

        if len(trues_this_run) == 0:
            continue

        acc = (np.array(preds_this_run) == np.array(trues_this_run)).mean() * 100.0
        all_accs.append(acc)
        all_preds_list.append(preds_this_run)
        all_trues_list.append(trues_this_run)
        all_metas_list.append(metas)

    if not all_accs:
        print("❌ 沒有有效測試結果。")
        return 0.0, 0.0, 0, 0.0

    avg_acc = float(np.mean(all_accs))
    max_acc = float(np.max(all_accs))
    min_acc = float(np.min(all_accs))
    std_acc = float(np.std(all_accs))
    best_index = int(np.argmax(all_accs))

    best_preds = all_preds_list[best_index]
    best_trues = all_trues_list[best_index]
    best_metas = all_metas_list[best_index]

    best_acc = float(all_accs[best_index])
    best_f1 = float(f1_score(best_trues, best_preds, average='macro'))
    cm = confusion_matrix(best_trues, best_preds)

    print(f"\n🏆 測試最佳準確率（第 {best_index+1} 次）：{best_acc:.2f}%")
    print(f"📊 對應 F1-score（macro）：{best_f1:.4f}")
    disp = ConfusionMatrixDisplay(cm, display_labels=list(label_mapping.keys()))
    disp.plot(cmap='Blues')
    plt.title(f"Best Confusion Matrix (Run {best_index+1})")
    plt.show()

    print(f"\n🎯 測試平均準確率（90 次）：{avg_acc:.2f}%")
    print("📊 最終統計結果（90 次抽樣）：")
    print(f"平均：{avg_acc:.2f}% | 最大：{max_acc:.2f}% | 最小：{min_acc:.2f}% | 標準差：{std_acc:.2f}%")

    # === 針對最佳那次：列出誤判清單 ===
    wrong_rows = []
    for t, p, m in zip(best_trues, best_preds, best_metas):
        if t != p:
            fname = os.path.basename(m["path"])
            wrong_rows.append({
                "person_id": m["person_id"],
                "file": fname,
                "true": inv_label_mapping.get(t, str(t)),
                "pred": inv_label_mapping.get(p, str(p))
            })

    if len(wrong_rows) == 0:
        print("\n✅ 最佳那次沒有誤判樣本。")
    else:
        wrong_rows.sort(key=lambda r: (str(r["person_id"]), r["true"], r["pred"], r["file"]))
        print("\n❌ 最佳那次誤判清單（person_id / 檔名 / 真實步態 → 預測步態）：")
        for r in wrong_rows:
            print(f" - person {r['person_id']:>2} | {r['file']:<20} | {r['true']:>6} → {r['pred']}")

        from collections import Counter as Cnt
        pair_counter = Cnt((r["true"], r["pred"]) for r in wrong_rows)
        print("\n📉 誤判對（真實→預測）次數統計：")
        for (tname, pname), cnt in pair_counter.most_common():
            print(f" - {tname:>6} → {pname:<6} : {cnt} 次")

    return best_acc, avg_acc, best_index + 1, best_f1

# === 單輪訓練（固定資料；加入「本輪最佳（最低平均 Loss）」保存與回傳） ===
def train_one_run_fixed(model, fixed_train_loader, device, total_epochs=200,
                        base_lr=0.01, weight_decay=0.001, step_size=40, gamma=0.1):
    # 類別權重
    y_tensor = fixed_train_loader.dataset.tensors[1].cpu().numpy().tolist()
    label_counter = Counter(y_tensor)
    total_samples = sum(label_counter.values())
    class_weights = []
    for i in range(num_classes):
        if i in label_counter and label_counter[i] > 0:
            class_weights.append(total_samples / label_counter[i])
        else:
            class_weights.append(1.0)
    class_weights = torch.tensor(class_weights, dtype=torch.float32, device=device)
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    # Adam（Adaptive Moment Estimation, ADAM）最佳化器
    optimizer = optim.Adam(model.parameters(), lr=base_lr, weight_decay=weight_decay)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=step_size, gamma=gamma)

    # 追蹤本輪最佳（以 epoch 平均 loss 最小為準）
    best_state_dict = None
    best_epoch = -1
    best_avg_loss = float('inf')

    model.train()
    for epoch in range(total_epochs):
        total_loss, correct, total = 0.0, 0, 0
        for X_batch, y_batch in fixed_train_loader:
            optimizer.zero_grad(set_to_none=True)
            X_in = X_batch.to(device).unsqueeze(-1)  # [N, C, V, T] → [N, C, V, T, 1]
            out = model(X_in)
            loss = criterion(out, y_batch.to(device))
            if torch.isnan(loss):
                continue
            loss.backward()
            optimizer.step()

            total_loss += float(loss.item())
            preds = out.argmax(dim=1)
            correct += int((preds.detach().cpu() == y_batch).sum().item())
            total += int(y_batch.size(0))

        scheduler.step()

        # 本 epoch 平均 loss（以 batch 數作分母）
        num_batches = max(len(fixed_train_loader), 1)
        epoch_avg_loss = total_loss / num_batches

        # 若更佳，保存一份權重（deepcopy）
        if epoch_avg_loss < best_avg_loss:
            best_avg_loss = epoch_avg_loss
            best_epoch = epoch + 1
            best_state_dict = deepcopy(model.state_dict())

        if (epoch + 1) % 5 == 0:
            train_acc = (correct / total * 100.0) if total > 0 else 0.0
            current_lr = scheduler.get_last_lr()[0]
            print(f"[Epoch {epoch+1:>4}] 🔧 LR: {current_lr:.6f} | AvgLoss: {epoch_avg_loss:.4f} | Train Acc: {train_acc:.2f}%")

    # 回傳本輪最佳（最低平均 Loss）權重與對應 epoch
    return best_state_dict, best_epoch, best_avg_loss

# === 50 次固定資料重複實驗（第 2 次之後接續前一輪最佳權重） ===
def run_mmgs_experiment_fixed_50():
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    # 取得資料與可用片段
    train_paths, train_labels, test_paths, test_labels = collect_paths()
    train_cache = preparse_valid_segments(train_paths)
    test_cache  = preparse_valid_segments(test_paths)

    print("📊 訓練集樣本數量分布：", Counter(train_labels))
    print("📊 測試集樣本數量分布：", Counter(test_labels))
    print(f"📁 原始測試影片數：{len(test_paths)}，有效影片數：{len(test_cache)}")

    results = []
    best_overall_avg = 0.0
    best_overall_state = None
    best_overall_f1 = 0.0
    best_overall_repeat = 0

    # 🔢 用來彙整 50 次的三個指標
    list_avg_acc_90 = []
    list_best_run_acc = []
    list_best_macro_f1 = []

    # Excel 輸出檔名
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    excel_path = f"results_mmgs_fixed_{ts}.xlsx"

    # 本輪最佳權重保存檔（供下一輪載入）
    last_run_ckpt_path = "last_run_best.pt"

    for repeat in range(1, 51):
        print(f"\n🔁 第 {repeat}/50 次實驗：固定抽樣 + 1200 epochs 訓練（第 2 次起接續上一輪最佳權重）")

        # 設置種子（控制抽樣與初始化）
        seed = 1000 + repeat
        set_seed(seed)

        # 固定抽樣一次（整輪訓練都用同一批）
        fixed_train_loader = make_fixed_train_loader(
            train_paths, train_labels, train_cache,
            clips_per_file=1, batch_size=32, seed=seed
        )
        if fixed_train_loader is None:
            print("⚠️ 本輪沒有有效的固定訓練資料，跳過。")
            continue

        # 建模
        model = build_ags_gcn().to(device)

        # ✅ 第 2 輪開始：若有上一輪最佳權重，先載入（接續訓練）
        if repeat > 1 and os.path.exists(last_run_ckpt_path):
            try:
                model.load_state_dict(torch.load(last_run_ckpt_path, map_location=device))
                print(f"↪️ 已接續上一輪最佳權重：{last_run_ckpt_path}")
            except Exception as e:
                print(f"⚠️ 載入上一輪權重失敗，改用隨機初始化：{e}")

        # 訓練（會回傳本輪最佳權重）
        best_state_dict, best_epoch, best_avg_loss = train_one_run_fixed(
            model, fixed_train_loader, device,
            total_epochs=1200, base_lr=0.01, weight_decay=0.001,
            step_size=400, gamma=0.1
        )

        # 若有更佳權重，套用之後再評估與保存
        if best_state_dict is not None:
            model.load_state_dict(best_state_dict)
            torch.save(best_state_dict, last_run_ckpt_path)
            print(f"💾 已保存本輪最佳權重（Epoch {best_epoch}, AvgLoss={best_avg_loss:.4f}）→ {last_run_ckpt_path}")
        else:
            # 若未記錄到最佳（理論上不會發生），就保存目前模型
            torch.save(model.state_dict(), last_run_ckpt_path)
            print(f"💾 未記錄到更佳權重，保存當前模型 → {last_run_ckpt_path}")

        # 測試（90 次）
        print("✅ 訓練完成，進行 90 次測試中...")
        best_acc, avg_acc, best_run_idx, best_f1 = evaluate_model_90_times(
            model, test_paths, test_labels, test_cache, device
        )

        # 紀錄指標到清單（用於 50 次統計）
        list_best_run_acc.append(best_acc)
        list_avg_acc_90.append(avg_acc)
        list_best_macro_f1.append(best_f1)

        # 記錄逐輪結果
        label_counts_fixed = Counter(fixed_train_loader.dataset.tensors[1].cpu().numpy().tolist())
        row = {
            "repeat": repeat,
            "seed": seed,
            "best_train_epoch": int(best_epoch if best_epoch != -1 else 0),
            "train_counts_knee": int(label_counts_fixed.get(label_mapping['knee'], 0)),
            "train_counts_normal": int(label_counts_fixed.get(label_mapping['normal'], 0)),
            "train_counts_padding": int(label_counts_fixed.get(label_mapping['padding'], 0)),
            "best_run_acc": float(best_acc),
            "avg_acc_90": float(avg_acc),
            "best_run_index": int(best_run_idx),
            "best_macro_f1": float(best_f1),
        }
        results.append(row)

        # 仍然追蹤「全局最佳」（Across 50 repeats，以 avg_acc 為主）
        if avg_acc > best_overall_avg:
            best_overall_avg = float(avg_acc)
            best_overall_state = deepcopy(model.state_dict())
            best_overall_f1 = float(best_f1)
            best_overall_repeat = int(repeat)
            torch.save(best_overall_state, "best_model_fixed.pt")
            print(f"🌟 更新全局最佳模型（第 {repeat} 輪）Avg: {avg_acc:.2f}% | Best-Run: {best_acc:.2f}% | F1: {best_f1:.4f} → 保存為 best_model_fixed.pt")

        # 追加寫入 Excel（每輪都更新一次 runs 分頁，避免中斷遺失）
        try:
            df_runs = pd.DataFrame(results)
            with pd.ExcelWriter(
                excel_path,
                engine="openpyxl",
                mode=("a" if os.path.exists(excel_path) else "w"),
                if_sheet_exists="replace"
            ) as writer:
                df_runs.to_excel(writer, index=False, sheet_name="runs")
            print(f"📝 已更新 Excel：{excel_path}")
        except Exception as e:
            print(f"⚠️ Excel 寫入失敗（不影響訓練）：{e}")

    # ===== 50 次的統計彙整 =====
    def stat_series(values):
        arr = np.array(values, dtype=float)
        return {
            "mean": float(np.mean(arr)) if arr.size else 0.0,
            "max":  float(np.max(arr))  if arr.size else 0.0,
            "min":  float(np.min(arr))  if arr.size else 0.0,
            "var":  float(np.var(arr))  if arr.size else 0.0,  # 變異數（母體變異數）
            "std":  float(np.std(arr))  if arr.size else 0.0,  # 標準差
            "n":    int(arr.size)
        }

    stats_avg_acc  = stat_series(list_avg_acc_90)
    stats_best_acc = stat_series(list_best_run_acc)
    stats_best_f1  = stat_series(list_best_macro_f1)

    # 彙整表格（寫入 Excel 的 summary 分頁）
    df_summary = pd.DataFrame([
        {"metric": "avg_acc_90(%)",   **stats_avg_acc},
        {"metric": "best_run_acc(%)", **stats_best_acc},
        {"metric": "best_macro_f1",   **stats_best_f1},
    ])[["metric", "n", "mean", "max", "min", "var", "std"]]

    # 顯示到終端
    print("\n================= 📈 50 次實驗統計（Across 50 repeats） =================")
    def pretty(d):
        return f"n={d['n']}, 平均={d['mean']:.4f}, 最大={d['max']:.4f}, 最小={d['min']:.4f}, 變異數={d['var']:.6f}, 標準差={d['std']:.6f}"
    print("avg_acc_90(%) :", pretty(stats_avg_acc))
    print("best_run_acc(%):", pretty(stats_best_acc))
    print("best_macro_f1  :", pretty(stats_best_f1))

    # Excel 寫入 summary 分頁
    try:
        with pd.ExcelWriter(
            excel_path,
            engine="openpyxl",
            mode=("a" if os.path.exists(excel_path) else "w"),
            if_sheet_exists="replace"
        ) as writer:
            # 先寫入 runs（保險起見再覆蓋一次最新結果）
            pd.DataFrame(results).to_excel(writer, index=False, sheet_name="runs")
            # 再寫入 summary
            df_summary.to_excel(writer, index=False, sheet_name="summary")
        print(f"📄 統計摘要已寫入 Excel 的 summary 分頁：{excel_path}")
    except Exception as e:
        print(f"⚠️ Excel summary 寫入失敗：{e}")

    # ===== 收尾輸出 =====
    print("\n✅ 50 輪實驗完成")
    print("📊 全局最佳摘要：")
    print(f"最佳平均準確率（90 次）：{best_overall_avg:.2f}%")
    print(f"對應 Macro F1-score：{best_overall_f1:.4f}")
    print(f"最佳模型來自第 {best_overall_repeat} 輪（best_model_fixed.pt）")
    print(f"📄 全部結果與統計已寫入：{excel_path}")
    print(f"🔁 若要接續上一輪，將從 {os.path.abspath('last_run_best.pt')} 載入。")

# === 直接執行 ===
if __name__ == "__main__":
    run_mmgs_experiment_fixed_50()
```

```python
# -*- coding: utf-8 -*-
# ✅ 改良版：同時計算最佳、最差與平均結果（含混淆矩陣、Macro F1、誤判清單）
# 適用於 best_model_fixed.pt 或其他相容模型

import os
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import confusion_matrix, f1_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from collections import Counter

from net.mt_gsgcn_model import Model

# === 使用者設定 ===
model_path = "best_model_fixed.pt"
data_dir = "../datasets/database_joints"
test_ids = [2, 20, 6, 11, 12, 13, 18, 26]
selected_joints = [18, 17, 16, 0, 12, 13, 14]
time_steps = 48
num_channels = 3
label_mapping = {'knee': 0, 'normal': 1, 'padding': 2}
inv_label_mapping = {v: k for k, v in label_mapping.items()}
NUM_RUNS = 90

# === 資料讀取 ===
def collect_paths():
    train_paths, train_labels, test_paths, test_labels = [], [], [], []
    for class_name, label in label_mapping.items():
        class_dir = os.path.join(data_dir, class_name)
        for person_id in os.listdir(class_dir):
            person_path = os.path.join(class_dir, person_id)
            if not os.path.isdir(person_path):
                continue
            pid = int(person_id)
            for file in os.listdir(person_path):
                if not file.endswith('.txt'):
                    continue
                fpath = os.path.join(person_path, file)
                if pid in test_ids:
                    test_paths.append(fpath)
                    test_labels.append(label)
                else:
                    train_paths.append(fpath)
                    train_labels.append(label)
    return train_paths, train_labels, test_paths, test_labels

def preparse_valid_segments(paths):
    cache = {}
    for path in tqdm(paths, desc="🔍 掃描測試片段"):
        raw_lines = []
        with open(path, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) == 202:
                    raw_lines.append(parts)
        if len(raw_lines) < time_steps:
            continue
        data = pd.DataFrame(raw_lines).astype(float)
        total_frames = len(data)
        valid_starts = list(range(0, total_frames - time_steps + 1))
        cache[path] = (data, valid_starts)
    return cache

def parse_person_id(path):
    return int(path.replace("\\", "/").split("/")[-2])

def sample_test_once(paths, labels, cache):
    X, y, metas = [], [], []
    for path, label in zip(paths, labels):
        if path not in cache: continue
        data, valid_starts = cache[path]
        if not valid_starts: continue
        start = np.random.choice(valid_starts)
        joints_xyz = np.zeros((data.shape[0], len(selected_joints), num_channels), dtype=np.float32)
        for i, j in enumerate(selected_joints):
            joints_xyz[:, i, 0] = data.iloc[:, 3 + j * 8]
            joints_xyz[:, i, 1] = data.iloc[:, 4 + j * 8]
            joints_xyz[:, i, 2] = data.iloc[:, 5 + j * 8]
        clip = joints_xyz[start:start + time_steps]
        if clip.shape[0] < time_steps: continue
        X.append(clip)
        y.append(label)
        metas.append({"path": path, "person_id": parse_person_id(path), "true_label": label})
    if not X: return None, None
    X = np.array(X).transpose(0, 3, 1, 2)
    y = np.array(y)
    loader = DataLoader(TensorDataset(torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.long)),
                        batch_size=32, shuffle=False)
    return loader, metas

def build_model():
    return Model(
        in_channels=3,
        num_class=3,
        graph_args={'layout': 'custom_7joints_spine', 'strategy': 'spatial'},
        edge_importance_weighting=True
    )

# === 評估函式（含誤判清單） ===
def evaluate_best_worst(model, device, test_paths, test_labels, test_cache, num_runs=90):
    model.eval()
    all_accs, all_preds, all_trues, all_metas = [], [], [], []

    for _ in range(num_runs):
        loader, metas = sample_test_once(test_paths, test_labels, test_cache)
        if loader is None: continue
        preds_run, trues_run = [], []
        with torch.no_grad():
            for Xb, yb in loader:
                X_in = Xb.to(device).unsqueeze(-1)
                logits = model(X_in)
                preds = logits.argmax(dim=1)
                preds_run.extend(preds.cpu().numpy().tolist())
                trues_run.extend(yb.numpy().tolist())
        acc = (np.array(preds_run) == np.array(trues_run)).mean() * 100
        all_accs.append(acc)
        all_preds.append(preds_run)
        all_trues.append(trues_run)
        all_metas.append(metas)

    # === 統計 ===
    avg_acc = np.mean(all_accs)
    best_idx, worst_idx = np.argmax(all_accs), np.argmin(all_accs)
    best_acc, worst_acc = all_accs[best_idx], all_accs[worst_idx]
    best_f1 = f1_score(all_trues[best_idx], all_preds[best_idx], average="macro")
    worst_f1 = f1_score(all_trues[worst_idx], all_preds[worst_idx], average="macro")
    avg_f1 = np.mean([f1_score(t, p, average="macro") for t, p in zip(all_trues, all_preds)])

    # ✅ 新增：準確率的樣本變異數與標準差
    var_acc = np.var(all_accs, ddof=1)  # 樣本變異數（建議）
    std_acc = np.sqrt(var_acc)          # 對應標準差
    # 若你想用母體變異數：var_acc = np.var(all_accs, ddof=0)

    print(f"\n🏆 最佳準確率：{best_acc:.2f}% | Macro F1：{best_f1:.4f}")
    print(f"📉 最差準確率：{worst_acc:.2f}% | Macro F1：{worst_f1:.4f}")
    print(f"📊 平均準確率：{avg_acc:.2f}% | 平均 Macro F1：{avg_f1:.4f}")
    print(f"📈 準確率變異數（樣本）：{var_acc:.6f} | 標準差：{std_acc:.6f}")
    # === 繪製混淆矩陣（分開顯示在輸出格）===
    cm_best = confusion_matrix(all_trues[best_idx], all_preds[best_idx])
    cm_worst = confusion_matrix(all_trues[worst_idx], all_preds[worst_idx])

    # ✅ Best Run 混淆矩陣（藍色）
    fig1, ax1 = plt.subplots(figsize=(4.5, 4))
    ConfusionMatrixDisplay(cm_best, display_labels=list(label_mapping.keys())).plot(
        cmap="Blues", ax=ax1, colorbar=False
    )
    ax1.set_title("")  # 移除標題
    plt.tight_layout()
    plt.show()  # ✅ 直接顯示在格子中

    # ✅ Worst Run 混淆矩陣（藍色，同樣風格）
    fig2, ax2 = plt.subplots(figsize=(4.5, 4))
    ConfusionMatrixDisplay(cm_worst, display_labels=list(label_mapping.keys())).plot(
        cmap="Blues", ax=ax2, colorbar=False
    )
    ax2.set_title("")  # 移除標題
    plt.tight_layout()
    plt.show()  # ✅ 直接顯示在格子中

    # === 誤判清單 ===
    def print_wrong_cases(idx, title):
        preds, trues, metas = all_preds[idx], all_trues[idx], all_metas[idx]
        wrong_rows = []
        for t, p, m in zip(trues, preds, metas):
            if t != p:
                fname = os.path.basename(m["path"])
                wrong_rows.append({
                    "person_id": m["person_id"],
                    "file": fname,
                    "true": inv_label_mapping[t],
                    "pred": inv_label_mapping[p]
                })
        print(f"\n❌ {title}（共 {len(wrong_rows)} 筆）:")
        if not wrong_rows:
            print("✅ 無誤判樣本")
            return
        wrong_rows.sort(key=lambda r: (r["person_id"], r["true"], r["pred"], r["file"]))
        for r in wrong_rows:
            print(f" - person {r['person_id']:>2} | {r['file']:<20} | {r['true']:>6} → {r['pred']}")
        # 統計誤判對
        pair_counter = Counter((r["true"], r["pred"]) for r in wrong_rows)
        print("\n📉 誤判對（真實→預測）次數統計：")
        for (tname, pname), cnt in pair_counter.most_common():
            print(f" - {tname:>6} → {pname:<6} : {cnt} 次")

    print_wrong_cases(best_idx, "最佳那次誤判清單")
    print_wrong_cases(worst_idx, "最差那次誤判清單")

    return best_acc, worst_acc, avg_acc, best_f1, worst_f1, avg_f1

# === 主程式 ===
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🖥 使用裝置：{device}")

_, _, test_paths, test_labels = collect_paths()
test_cache = preparse_valid_segments(test_paths)
print(f"📁 測試影片：{len(test_paths)}，有效影片：{len(test_cache)}")

model = build_model().to(device)
model.load_state_dict(torch.load(model_path, map_location=device))
print(f"✅ 已載入模型：{model_path}")

evaluate_best_worst(model, device, test_paths, test_labels, test_cache, num_runs=NUM_RUNS)
```
